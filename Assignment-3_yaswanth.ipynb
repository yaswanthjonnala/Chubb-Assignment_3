**1.	Ingest sample order data into a Spark DataFrame.**

```python
from pyspark.sql import functions as F
from delta.tables import DeltaTable
#loading the data set
df_raw = (
    spark.read
        .option("header", "true")
        .option("inferSchema", "true")
        .csv("/Volumes/yaswanth/default/orders/orders_5000.csv")
)

df_raw.show(5)
df_raw.printSchema()

```

**2.	Add a derived column order_date (date only from order_timestamp).**

```python
# Adding order_date derived from order_timestamp
df_orders = df_raw.withColumn(
    "order_date",
    F.to_date("order_timestamp")   # keep only the date part
)

# verifying the new order_date column
display(df_orders.select("order_id", "order_timestamp", "order_date", "country", "amount"))

```

**3.	Write the DataFrame as a Delta table partitioned by country and order_date.**

```python
#  Write DataFrame as a Delta table with partitioning
(
    df_orders.write
        .format("delta")
        .mode("overwrite")                    # overwrite for first run
        .partitionBy("country", "order_date") # partition columns
        .save("/Volumes/yaswanth/default/orders/orders_delta")
)
```

```python

# This loads all written Delta files so we can continue queries, updates, and other tasks.
df_delta = spark.read.format("delta").load(
    "/Volumes/yaswanth/default/orders/orders_delta"
)

# Displaying the data to confirm the Delta table loaded correctly
display(df_delta)
```

**6.	Demonstrate Delta Lake Time Travel:Write data, update some rows, then query older versions.**

```python
from delta.tables import DeltaTable

# View table history using the Delta path
delta_tbl = DeltaTable.forPath(
    spark, "/Volumes/yaswanth/default/orders/orders_delta"
)

history_df = delta_tbl.history()  # get full history

display(history_df)

```

```python
# Perform an UPDATE to create a new Delta version
delta_tbl.update(
    condition="status = 'CREATED' AND amount > 300",
    set={"status": F.lit("PAID")}
)

# Show updated history (new version should appear)
history_after_update = delta_tbl.history()
display(history_after_update)
```

```python
# Read older version vs latest version
# We load by Delta path using versionAsOf

# Load version 0 (initial dataset)
df_v0 = (
    spark.read.format("delta")
         .option("versionAsOf", 0)
         .load("/Volumes/yaswanth/default/orders/orders_delta")
)

# Load latest Delta version
df_latest = spark.read.format("delta").load(
    "/Volumes/yaswanth/default/orders/orders_delta"
)

#  status counts for version 0
print("Status counts in Version 0 (before update):")
display(df_v0.groupBy("status").count())

#  status counts for the latest version
print("Status counts in Latest Version (after update):")
display(df_latest.groupBy("status").count())
```

**7.	Demonstrate Schema Evolution:
o	Add payment_method & coupon_code to new data.
o	Write to the same Delta table, allowing schema evolution.**


```python
# Creating new batch with extra columns
# Simulating new incoming data with payment_method and coupon_code
df_new = (
    df_delta.limit(1000)  # takes sample rows as "new data"
        .withColumn(
            "payment_method",
            F.when(F.col("country") == "IN", "UPI")   # simple rule
             .when(F.col("country").isin("US", "CA"), "CARD")
             .otherwise("COD")
        )
        .withColumn(
            "coupon_code",
            F.when(F.col("amount") > 400, "BIGSAVE40")
             .when(F.col("amount") > 200, "SAVE20")
             .otherwise(F.lit(None).cast("string"))
        )
)

# Displaying sample of new data with added columns
display(df_new.select("order_id", "country", "amount", "payment_method", "coupon_code").limit(10))
```

```python
#  Append new data and let Delta automatically
# evolve schema using mergeSchema = true
(
    df_new.write
        .format("delta")
        .mode("append")                   # add to existing Delta table
        .option("mergeSchema", "true")    # allow new columns
        .save("/Volumes/yaswanth/default/orders/orders_delta")
)
```

```python
#  Check schema of the Delta table after evolution
schema_after = spark.sql("""
    DESCRIBE DETAIL delta.`/Volumes/yaswanth/default/orders/orders_delta`
""")
```

```python

# Read the Delta table and filter rows where
# new columns (payment_method) are populated

df_after = spark.read.format("delta").load(
    "/Volumes/yaswanth/default/orders/orders_delta"
)

result_new_cols = (
    df_after
        .filter(F.col("payment_method").isNotNull())
        .select("order_id", "country", "amount", 
                "payment_method", "coupon_code")
        .limit(10)
)

display(result_new_cols)
```

**8.	Demonstrate Updates & Deletes using Delta:
o	Mark some orders as CANCELLED.
o	Delete orders below a certain amount (e.g., test data cleanup).**


```python
# Updating rows using DeltaTable.forPath()
# Business rule: amount between 5 and 20 => CANCELLED

from delta.tables import DeltaTable

delta_tbl = DeltaTable.forPath(
    spark, "/Volumes/yaswanth/default/orders/orders_delta"
)

delta_tbl.update(
    condition="amount >= 5 AND amount < 20",
    set={"status": F.lit("CANCELLED")}
)
```

```python
# Read Delta table and compute status counts

df_after_update = spark.read.format("delta").load(
    "/Volumes/yaswanth/default/orders/orders_delta"
)

status_counts_after_update = (
    df_after_update
        .groupBy("status")
        .count()
)

display(status_counts_after_update)
```

```python
# Delete rows (amount < 5)
delta_tbl.delete("amount < 5")
```

```python
# Count rows again after delete

df_after_delete = spark.read.format("delta").load(
    "/Volumes/yaswanth/default/orders/orders_delta"
)

display(df_after_delete.count())
```

**10.	(Bonus) Show how small file problems can occur with too many partitions and how OPTIMIZE helps.**

```python

# Creating a separate demo table with MANY tiny files
# This intentionally causes "small-file problem"

volume_small_path = "/Volumes/yaswanth/default/orders/orders_smallfiles"

# Create many tiny delta files
for i in range(100):                   # 100 tiny writes
    tiny_df = df_orders.limit(50)      # very small batch
    (tiny_df.write
        .format("delta")
        .mode("append")
        .save(volume_small_path))

```

```python

# Listing files BEFORE OPTIMIZE to show fragmentation
files_before = dbutils.fs.ls(volume_small_path)
display(files_before)
print("Number of files before OPTIMIZE:", len(files_before))
```

```python
# Optimizing to compact many small files into bigger ones
opt_result = spark.sql(f"""
    OPTIMIZE delta.`{volume_small_path}`
""")
display(opt_result)
```
